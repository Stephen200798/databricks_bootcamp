#BUSINESS TRANSFORMATION AND MODELING

```python
%sql
SELECT 
    sd.Sales_Order_Num,
    pr.Product_number,
    cu.Customer_number,
    sd.Sales_Order_Date,
    sd.Sales_Ship_Date,
    sd.Sales_Due_Date,
    sd.Sales_Sales AS Sales,
    sd.Sales_Quantity,
    sd.Sales_Price
FROM silver.crm_sales_details sd
LEFT JOIN gold.dim_customers cu  
ON sd.Sales_Customer_ID = cu.Customer_id
LEFT JOIN gold.dim_products pr  
ON sd.Sales_Product_Key = pr.Product_key
```

#WRITE INTO GOLD FACT TABLE

```python
%sql
DROP VIEW IF EXISTS gold.fact_sales
```

```python
%sql
CREATE VIEW gold.fact_sales AS
SELECT 
    sd.Sales_Order_Num,
    pr.Product_number,
    cu.Customer_number,
    sd.Sales_Order_Date,
    sd.Sales_Ship_Date,
    sd.Sales_Due_Date,
    sd.Sales_Sales AS Sales,
    sd.Sales_Quantity,
    sd.Sales_Price
FROM silver.crm_sales_details sd
LEFT JOIN gold.dim_customers cu  
ON sd.Sales_Customer_ID = cu.Customer_id
LEFT JOIN gold.dim_products pr  
ON sd.Sales_Product_Key = pr.Product_key
```
