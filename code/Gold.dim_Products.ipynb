#BUSINESS TRANSFORMATION AND MODELING

```python
%sql
SELECT
  ROW_NUMBER() OVER (ORDER BY pr.Product_start_dt, pr.Product_key) AS Product_number,
  pr.Product_id,
  pr.Product_key,
  pr.Product_name,
  pr.Category_id,
  px.Category,
  px.Subcategory,
  pr.Product_cost,
  pr.Product_line,
  px.Maintenance,
  pr.Product_start_dt
FROM silver.crm_prd_info pr
LEFT JOIN silver.erp_px_cat_g1v2 px
ON pr.Category_id = px.Customer_id
WHERE Product_end_dt IS NULL
```

#WRITE INTO GOLD TABLE

```python
%sql
DROP VIEW IF EXISTS gold.dim_Products
```

```python
%sql
CREATE VIEW IF NOT EXISTS gold.dim_Products AS 
SELECT
  ROW_NUMBER() OVER (ORDER BY pr.Product_start_dt, pr.Product_key) AS Product_number,
  pr.Product_id,
  pr.Product_key,
  pr.Product_name,
  pr.Category_id,
  px.Category,
  px.Subcategory,
  pr.Product_cost,
  pr.Product_line,
  px.Maintenance,
  pr.Product_start_dt
FROM silver.crm_prd_info pr
LEFT JOIN silver.erp_px_cat_g1v2 px
ON pr.Category_id = px.Customer_id
WHERE Product_end_dt IS NULL
```
